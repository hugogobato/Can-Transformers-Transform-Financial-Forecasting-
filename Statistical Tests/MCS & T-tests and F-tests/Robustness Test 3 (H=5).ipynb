```python
import pandas as pd
import numpy as np
```

```python
import numpy as np
import scipy

def f_test(x, y):
    x = np.array(x)
    y = np.array(y)
    variance1=np.var(x, ddof=1)
    variance2=np.var(y, ddof=1)
    # Ensuring the larger variance is the numerator
    if variance1 > variance2:
      f_value = variance1 / variance2
      df1 = len(x) - 1  # degrees of freedom for x
      df2 = len(y) - 1  # degrees of freedom for y
    else:
      f_value = variance2 / variance1
      df1 = len(y) - 1  # degrees of freedom for y
      df2 = len(x) - 1  # degrees of freedom for x
    p = 1-scipy.stats.f.cdf(f_value, df1, df2) #find p-value of F test statistic
    return f_value, p
```

```python
NBEATSx=pd.read_excel("NBEATSx forecasts.xlsx", index_col="Unnamed: 0")
```

```python
NHITS=pd.read_excel("NHITS forecasts.xlsx", index_col="Unnamed: 0")
```

```python
TFT=pd.read_excel("TFT forecasts.xlsx", index_col="Unnamed: 0")
```

```python
TimesNet=pd.read_excel("TimesNet forecasts.xlsx", index_col="Unnamed: 0")
```

```python
Autoformer=pd.read_excel("Autoformer forecasts.xlsx", index_col="Unnamed: 0")
```

```python
Informer=pd.read_excel("Informer forecasts.xlsx", index_col="Unnamed: 0")
```

```python
PatchTST=pd.read_excel("PatchTST forecasts.xlsx", index_col="Unnamed: 0")
```

# Results

```python
models = [NHITS,NBEATSx,TFT, TimesNet, Autoformer, Informer, PatchTST]
models_names = ["NHITS","NBEATSx","TFT", "TimesNet", "Autoformer", "Informer", "PatchTST "]
for j in range(len(models)):
  print(f"{models_names[j]}")
  for i in range(len(NBEATSx.columns)):
    print(f"Mean of {models[j].columns[i]}: {round(100*models[j][models[j].columns[i]].mean(),3)}%")
    print(f"Standard Deviation of {models[j].columns[i]}: {round(100*models[j][models[j].columns[i]].std(),3)}%")
```

# Autoformer

T-tests

```python
from scipy import stats
models = [NHITS,NBEATSx,TFT, TimesNet, Informer, PatchTST]
models_names = ["NHITS","NBEATSx","TFT", "TimesNet", "Informer", "PatchTST "]
for j in range(len(models)):
  print(f"{models_names[j]}")
  for i in range(len(NBEATSx.columns)):
    print(f"{models[j].columns[i]}: {stats.ttest_ind(models[j][models[j].columns[i]], Autoformer[models[j].columns[i]])}")
```

F-tests

```python
models = [NHITS,NBEATSx,TFT, TimesNet, Informer, PatchTST]
models_names = ["NHITS","NBEATSx","TFT", "TimesNet", "Informer", "PatchTST "]
for j in range(len(models)):
  print(f"{models_names[j]}")
  for i in range(len(NBEATSx.columns)):
    print(f"{models[j].columns[i]}: {f_test(models[j][models[j].columns[i]], Autoformer[models[j].columns[i]])}")
```
